```python
import numpy as np
import pandas as pd
```

```python
dataset1 = pd.read_csv('final_minkyu.txt', delimiter='\t', encoding='utf-8')
dataset2 = pd.read_csv('final_baekjun.txt', delimiter='\t', encoding='utf-8')
dataset3 = pd.read_csv('final_sangjun.txt', delimiter='\t', encoding='utf-8')
dataset4 = pd.read_csv('final_woohyun.txt', delimiter='\t', encoding='utf-8')
dataset5 = pd.read_csv('final_seungji.txt', delimiter='\t', encoding='euc-kr')
```

```python
del dataset3['Unnamed: 0']
del dataset4['Unnamed: 0']
del dataset5['Unnamed: 0']
```

```python
dataset1.shape, dataset2.shape, dataset3.shape, dataset4.shape, dataset5.shape
```

```python
dataset12 = pd.merge(dataset1, dataset2, how='inner')
dataset124 = pd.merge(dataset12, dataset4, how='inner')
dataset1245 = pd.merge(dataset124, dataset5, how='inner')
dataset = pd.merge(dataset3, dataset1245, how='left')
```

```python
dataset
```

```python
dataset.to_csv('dataset_final.csv', index=False, sep='\t', encoding='utf-8')
```
